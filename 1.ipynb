```python
import os
from dotenv import load_dotenv 
load_dotenv()
GOOGLE_API_KEY=os.getenv("GOOGLE_API_KEY")
```

```python
import google.generativeai as genai 
genai.configure()
```

```python
###displying the all the available models in the google gemini ai 
for model in genai.list_models(): 
    print((model.name))
```

```python
###gettting the specific model 
model=genai.GenerativeModel("gemini-pro")
```

```python
##generate the content 
contnet=model.generate_content("tell me about the india in the 10 lines ")
contnet.text
```

```python
##generating the content like streaming 
response=model.generate_content("give me future technologies with the great scop",stream=True)
for chunk in response: 
    print(chunk.text)
    print("_")
```
